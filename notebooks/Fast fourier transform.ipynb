```python
import os
import csv
import librosa
import librosa.display
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 200
f = 0.01
fs = 0.3
dt = 1/fs
x = np.arange(0,T,dt)
y = np.sin(2*np.pi*f*x)
plt.figure(figsize=(15,5))
plt.stem(x,y,linefmt='C1--',markerfmt='kd')
x_sig = np.arange(0,T,0.001)
y_sig = np.sin(2*np.pi*f*x_sig)
plt.plot(x_sig,y_sig,'k')
plt.xticks(np.arange(0,T,1/f))
plt.ylabel('amplitude')
plt.title('freqency f={} hz,sampling freqency={} hz,period ={} sec,duration={}sec'.format(f,fs,1/f,T))
_=plt.xlabel('time (sec)')
```

```python
D = 2/len(y)*np.abs(np.fft.fft(y))
fr = np.fft.fftfreq(len(y),d=dt)
mask = fr>0
plt.figure(figsize=(15,5))
plt.plot(fr[mask],D[mask],'k')
plt.title('FFT plot')
plt.ylabel('amplitude')
_=plt.xlabel('frequency')
```
